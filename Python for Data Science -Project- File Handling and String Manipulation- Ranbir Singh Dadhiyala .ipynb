```python
# opening the project file. using open() function with "r" mode to open and read the given txt file
project_file = open("conv.txt", "r")
```

```python
# reading the project file, using readlines() function to read the file line by line and storing the text in "lines" variable.
lines = project_file.readlines()
```

```python
# importing re for regular expressions opertions for special characters removal 
import re
```

```python
lines
```

```python
# lines is a list type variable
type(lines)
```

```python
# Removing "\n" new line character, as it is a blank line, no need to read it.
# using "for" loop to iterate over the "lines"
for line in lines:
    if line !="\n":
        print(line)
```

```python
type(line)
```

```python
# now we just need the text/ the words, we need to remove the colon ":" which acts as a splitter here.
# we are using "speakers" for names of dialogue speakers, which are on the left side of colon(splitter here)
# and "dialogue" for the dialogues spoken by the dialogue speakers, which are on the right of colon(splitter here)
# [0] will be used for left side of the splitter as index 0 
# [1] will be used for the right side of the splitter as index 1
# The split() function splits a string into a list. You can specify the separator, default separator is any whitespace. 
# we will change the words(dialogues) in lowercase so that there is no repetition of words like "The" and "the" 
# and we will get the unique words.
# imported re for regular expressions opertions for special characters removal using re.sub()
# using replace() function to replace some words with other words to avoid repetition 
for line in lines:
    if line !="\n":
        line = line.split(":")
        speakers = line[0]
        dialogue = line[1]
        dialogue = dialogue.lower()
        dialogue_spoken = re.sub('[^a-zA-Z0-9\n\.]', ' ', dialogue)
        dialogue_spoken = dialogue_spoken.replace(".", "").replace("yours", "your")
        dialogue_spoken = dialogue_spoken.replace("  ", " ").replace("   ", " ").replace("he s ", "he ").replace("deserter  if", "deserter if").replace("she s ", "she ").replace("the'seven", "the seven").replace("where s ", "where ").replace("where's ", "where ").replace("that s ", "that ").replace("that's ", "that ")
        dialogue_spoken = dialogue_spoken.replace(" ve", "'ve").replace("won t", "won't").replace("it's", "it").replace("it s", "it").replace("don t", "don't").replace(" ll", "'ll").replace("haven t", "haven't")
        dialogue_spoken = dialogue_spoken.replace("father s", "father").replace("mother s", "mother").replace("didn t", "didn't").replace("night s", "night ").replace(" m ", "'m ").replace(" s ", "").replace("he'soff","he's off").replace("me  you","me you")
        print(speakers, dialogue_spoken)
```

```python
# we will create an empty set and then store no. of speakers with the respective names in the sample conversation in empty set
# here empty set is named as speakers_set.
# we will use set because, In sets, there is no duplicate and also we need the unique count.
# we will also see the length, using length function len() to know the no. of speakers.
# we will use add() function, to add the list of speakers in the empty set (speakers set).
speakers_set = set()
for line in lines:
    if line !="\n":
        line = line.split(":")
        speakers = line[0]
        dialogue = line[1]
    speakers_set.add(speakers)
print(" The number of unique dialogue speakers/ unique speakers in the sample conversation are: ", len(speakers_set) )
print(speakers_set)
```

```python
type(speakers_set)
```

```python
# we created an empty dictionary here named simply as dictionary, to store the names of the speakers with the empty sets.
# so we can store the words spoken by them in the set later.
# we used set here as it has a property that it doesn't repeat elements.
dictionary = {}

for speakers in speakers_set:
    dictionary[speakers] = set()
```

```python
# in the above empty sets of the speakers we will fill the dialogues (words spoken => dialogues) 
# by them in the sample conversation 
# word by word as a dictionary of sets.
# as it is a dictionary of sets and sets has only the non- duplicate unique elements (here words)
# we will update the dictionary (using update() function) with the dialogue splitted word by word (using split() function) in it
# as a nested function.
for line in lines:
    if line != "\n":
        line = line.split(":")
        speakers = line[0]
        dialogue = line[1]
        dialogue = dialogue.lower()
        dialogue_spoken = re.sub('[^a-zA-Z0-9\n\.]', ' ', dialogue)
        dialogue_spoken = dialogue_spoken.replace(".", "").replace("yours", "your")
        dialogue_spoken = dialogue_spoken.replace("  ", " ").replace("   ", " ").replace("he s ", "he ").replace("deserter  if", "deserter if").replace("she s ", "she ").replace("the'seven", "the seven").replace("where s ", "where ").replace("where's ", "where ").replace("that s ", "that ").replace("that's ", "that ")
        dialogue_spoken = dialogue_spoken.replace(" ve", "'ve").replace("won t", "won't").replace("it's", "it").replace("it s", "it").replace("don t", "don't").replace(" ll", "'ll").replace("haven t", "haven't")
        dialogue_spoken = dialogue_spoken.replace("father s", "father").replace("mother s", "mother").replace("didn t", "didn't").replace("night s", "night ").replace(" m ", "'m ").replace(" s ", "").replace("he'soff","he's off").replace("me  you","me you")
        dictionary[speakers].update(dialogue_spoken.split())
        print(speakers, dialogue_spoken)
```

```python
# we are now writing/saving the file for individual speakers with the respective dialogues using "w" mode for write.
# we have used the speakers name as their respective file names (txt files).
# writing dialogues in 17 (unique no. of dialogue speakers/ speakers) different txt files word by word 
# which will be unrepeated words spoken by them.
# all individual files for individual dialogue speakers with the word by word dialogue(words spoken) by them.
# Here we are using writelines() function to write the line by line
# writelines() function is used to write a list of strings in an opened file.
# we are using join() function to concatenate the "\n" new line character with each word in the list_of_words
# list_of_words, which is the list of unique (unrepeated) words spoken by the speakers
for speakers in dictionary.keys():
    file = open(str(speakers) + ".txt" , "w")
    list_of_words = list(dictionary[speakers])
    file.writelines("\n".join(list_of_words))
    file.close()
```
