# Representation code:

Please Note: As the data in the two files have same uniq id, we can use to concatinate on the id to get the required representation.

```python
# Making the Dataframes for required columns for the required representation
result_flipkart_df = flipkart_df[["product_name", 'retail_price', "discounted_price"]]
result_amz_df = amz_df[["product_name", 'retail_price', "discounted_price"]]
```

```python
# Adding a suffix in the Flipkart and Amazon Dataframes required columns.
result_flipkart_df.columns = [ i+" in Flipkart"for i in result_flipkart_df.columns]
result_amz_df.columns = [ i+" in Amazon"for i in result_amz_df.columns]
```

```python
result_flipkart_df.head()
```

```python
result_amz_df.head()
```

```python
# concatenating both the data frames for the required representation.
result = pd.concat([result_flipkart_df, result_amz_df], axis="columns")
```
